### Problem statement

Let us consider a standard, shuffled deck of cards, defined as a list below.

```python
cards_deck = ['3 of Spades', 'Jack of Diamonds', '2 of Hearts', '4 of Clubs', '8 of Clubs', \
              '4 of Spades', '9 of Hearts', 'Jack of Spades', '2 of Diamonds', '9 of Spades', \
              'Ace of Hearts', 'Jack of Hearts', '4 of Hearts', 'Queen of Clubs', '9 of Clubs', \
              'King of Hearts', 'Ace of Clubs', 'Ace of Diamonds', '8 of Diamonds', 'King of Diamonds', \
              '6 of Diamonds', '7 of Spades', 'King of Clubs', '10 of Clubs', '9 of Diamonds', \
              '3 of Hearts', '7 of Diamonds', '6 of Clubs', '4 of Diamonds', '3 of Diamonds', \
              '6 of Spades', '5 of Hearts', 'Jack of Clubs', 'King of Spades', '5 of Diamonds', \
              '10 of Spades', 'Queen of Hearts', '2 of Clubs', '2 of Spades', 'Ace of Spades', \
              'Queen of Spades', '5 of Clubs', '7 of Clubs', '6 of Hearts', '10 of Diamonds', \
              '10 of Hearts', 'Queen of Diamonds', '5 of Spades', '3 of Clubs', '7 of Hearts', \
              '8 of Hearts', '8 of Spades']
print(cards_deck)
```

Let us now consider we have $ n $ players.
- The value of $ n $ is an integer greater or equal to 1, and lower than 8.
- The players are referred to as player $0$, player $1$, ..., player $(n-1)$.
- Players are seating around a table in a clockwise manner: player $ 1 $ is seating on the left of player $ 0 $, player $ 2 $ is seating on the left of player $ 1 $, etc.

### Task

Write a function **distribute_cards()**, which
- receives a shuffled deck of cards (such as the one above), *cards_deck*, as its first input parameter,
- receives a number of players seating at the table, *players_number*, as its second input parameter,
- receives an integer, *player_index*, with value in $ [0, $ players_number $ - 1] $, as its third input parameter. This corresponds to a specific player sitting at the table. For instance, we could pretend that this is the position you are currently sitting at!
- receives an integer, *start_from*, with value in $ [0, $ players_number $ - 1] $, as its fourth input parameter. This value corresponds to the first player to receive a card.

The function **distribute_cards()** then 
- distributes all cards in the deck,
- one card at a time,
- in clockwise order,
- between all players,
- starting from the player with index *start_from*.
- It is ok if some players have one more card than other players (for instance, if the number of cards is not divisible by number of players)

The function **distribute_cards()** returns a list of cards, *player_hand*, corresponding to the hand of player with index *player_index*, after we are done distributing the cards.

### Example



For instance, if we consider the shuffled deck defined above, $ 4 $ players and start distributing from player $ 0 $, player $ 0 $'s hand will be as given below.

```python
player_hand = ['3 of Spades', '8 of Clubs', '2 of Diamonds', '4 of Hearts', 'Ace of Clubs', \
               '6 of Diamonds', '9 of Diamonds', '4 of Diamonds', 'Jack of Clubs', 'Queen of Hearts', \
               'Queen of Spades', '10 of Diamonds', '3 of Clubs']
print("Deck: ", cards_deck)
print("---")
print("Player 0 hand: ", player_hand)
```

### Important note before we start

This question does not require anything super complex in terms of code. Typically, we will NOT implement a for/while loop for this task. In fact, the entire function, can be solved by using a smart **slicing** on the deck of cards. This means our function should simply consists of **cards_deck[a:b:c]** with some (nicely chosen) values **a**, **b** and **c** that you need to figure out.

**In other words, this means that this activity probably requires some work on paper to figure out the correct slicing parameters!**

### Your code below!

**Hint:** The difficult part is to identify what the index of the first card will be. Keep in mind that the *player_index* value, might be lower or higher than the *start_from* value. Your code should still work, in both cases!

```python
def distribute_cards(cards_deck, players_number, player_index, start_from):
    if(player_index >= start_from):
        first_index = player_index - start_from
    else:
        first_index = player_index - start_from + players_number
    player_hand = cards_deck[first_index::players_number]
    return player_hand
```

```python
def distribute_cards(cards_deck, players_number, player_index, start_from):
    first_card = ((player_index - start_from) % players_number)
    total_cards = len(cards_deck)
    player_hand = cards_deck[first_card:total_cards:players_number]
    return player_hand
```
