# Activity 1 - How many items in my inventory - solution

### Task

In several video games, the main character will have an inventory, i.e. a list of items that he/she is carrying at the moment.



This inventory could be defined as a list, as shown below.

```python
inventory1 = ["Sword", "Armor", "Potion", "Potion", "Torch", "Potion", "Bow", "Potion", "Torch", "Potion"]
print(inventory1)
```

Our objective is to write a function **how_many_items()**, which:
- receives an inventory list, such as the one above, as its first parameter,
- receives an item name, as a second parameter (for instance *item_name = "Torch"*)
- and returns the number of times the item in question appears in the inventory.

### Your code below!

```python
def how_many_items(inventory, item_name):
    counter = 0
    for item in inventory:
        if item == item_name:
            counter += 1
    return counter
```

### Expected results

```python
# This should return 5, as we have 5 potions in the inventory.
inventory1 = ["Sword", "Armor", "Potion", "Potion", "Torch", "Potion", "Bow", "Potion", "Torch", "Potion"]
print(how_many_items(inventory = inventory1, item_name = "Potion"))
```
