### Tasks

In many games, designers have implemented an experience/level system, where you can gain levels after having gained enough experience points. In general, the amount of experience required increases after each level.



In our game, we are going to claim that the amount of experience needed to go:
- from level 1 to 2 is equal to 1 exp point,
- from level 2 to 3 is equal to 4 exp point,
- from level 3 to 4 is equal to 9 exp point,
- from level $ n $ to $ n + 1 $ is equal to $ n^2 $ exp point.

Your objective is then to write a function **exp_to_level()** that
- receives a strinctly positive integer number *n* as its only parameter,
- returns the total ammount of experience required to get to level $ n + 1 $, defined as the *sum of squared numbers* from *1* to *n* as a result.

For instance, exp_to_level(3), will return $ 1^2 + 2^2 + 3^2 = 14 $ as a result.

### Your code below!

```python
def exp_to_level(n):
    
    counter = 0
    total = 0
    while(counter <= n):
        total += counter**2
        counter += 1
        
    return total
```
